### Polynomial Regression
Si sus puntos de datos claramente no se ajustan a una regresión lineal (una línea recta a través de todos los puntos de datos), podría ser ideal para la regresión polinomial.

La regresión polinomial, como la regresión lineal, usa la relación entre las variables xey para encontrar la mejor manera de trazar una línea a través de los puntos de datos.

```
import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

plt.scatter(x, y)
plt.show()
```

```
import numpy
import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

myline = numpy.linspace(1, 22, 100)
plt.style.use('classic')
plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```
import pandas as pd 
df = pd.read_csv("/content/drive/MyDrive/fh_5yrs.csv")
```

```
# Filtro los datos de Apple en el dataframe original
filter_apple = df['symbol'] == 'AAPL'
# Creo un nuevo dataframe llamado df_apple y almaceno todos los datos verdaderos del filtro que se aplico
df_apple = df[filter_apple]
# Finalmento imprimo en consola las 5 primeras filas
df_apple.head(100)
```

Extraemos 75 datos

```
df_s = df_apple.head(75)
```

```
y = df_s.volume
x = df_s.high
```

### Graficación de los datos: Precio de apertura = x y Volumen = y

```
fig, ax = plt.subplots(figsize=(20,7))
plt.scatter(x, y)
plt.show()
```

¿Funciona una regresión lineal?

```
mymodel = numpy.poly1d(numpy.polyfit(x, y, 1))
plt.style.use('classic')
myline = numpy.linspace(200, 400, 400)
fig, ax = plt.subplots(figsize=(20,7))
plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.grid(axis='both')
plt.show()
```

### Ajuste del modelo

```
mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))
plt.style.use('classic')
myline = numpy.linspace(220, 370, 50)
fig, ax = plt.subplots(figsize=(12,10))
plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.grid(axis='both')
plt.show()
```
