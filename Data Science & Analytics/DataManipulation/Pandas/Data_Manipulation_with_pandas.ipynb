1.2 Manipular dataframe

```
movies.sort_values('duration') #ordena los datos insteresantes
```

```
movies.sort_values("duration", ascending=True) #Ascendente
```

```
movies.sort_values("duration", ascending=False) #Descendente
```

```
movies[movies['num_critic_for_reviews'] > 400] #Filtrar con operaciones logicas.
```

```
movies[movies['director_name'] == 'James Cameron'] #Comparaciones directas
```

1.2.1 Solicitudes múltiples.

```
movies = pd.read_csv('movies2.csv')
duracion = movies['director_name'] == 'Gore Verbinski'
critic = movies['actor_2_name'] == 'Orlando Bloom'
movies[duracion & critic]
```

```
directores = movies['director_name'].isin(['Gore Verbinski', 'James Cameron']) #Metodo isin para varios filtros en una variable.
movies[directores]
```

```
movies
```

```
ascending_director_by_critics = movies.sort_values(["num_critic_for_reviews", "director_name"], ascending=[False, True])
ascending_director_by_critics
```
