## Getting started

```
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

### Exaple 1: Scatter plot

```
height = [62, 64, 69, 75, 66, 68, 65, 71, 76, 73]
weight = [120, 136, 148, 175, 137,165, 154, 172, 200, 187]
sns.scatterplot(x=height, y=weight)
plt.show()
```

### Exaple 2: Count plot

```
import seaborn as sns
import matplotlib.pyplot as plt
gender = ["Female", "Female", "Female", "Female", "Male", "Male", "Male", "Male", "Male", "Male"]
sns.countplot(x=gender)
plt.show()
```

### Using Pandas with Seaborn

```
import pandas as pd
df = pd.read_csv('countries-of-the-world.csv')
sns.countplot(data = df, y = 'Region')
```

### Scatterplot simple

```
df = pd.read_csv('tips.csv')
sns.scatterplot(x='total_bill', y='tip',data=df, hue='smoker', hue_order=['Yes','No'])
plt.show()
```

### Scatterplot personalizado

image.png

```
hue_colors = {'Yes':'Black',
              'No':'Red'}
sns.scatterplot(x='total_bill', y='tip',data=df, hue='smoker', palette=hue_colors)
plt.show()
```

```
hue_colors = {"Yes": "#808080", "No": "#00FF00"}
sns.scatterplot(x="total_bill", y="tip", data=df, hue="smoker", palette=hue_colors)
plt.show()
```

### Using hue with count plots

```
palette_colors = {"Female": "green", "Male": "blue"}
sns.countplot(x="smoker",data=df,hue="sex", palette=palette_colors)
plt.show()
```

### Introduction torelational plots andsubplots

Questions about quantitative variables
Relational plots
* Height vs. weight
* Number of school absences vs. nal grade
* GDP vs. percent literate


Introducing relplot()
* Create "relational plots": scatter plots or line plots
Why use relplot() instead of scatterplot()?
* relplot() lets you create subplots in a single gure


```
sns.relplot(x="total_bill",y="tip",data=df,kind="scatter")
plt.show()
```

```
sns.relplot(x="total_bill", y="tip", data=df, kind="scatter",col="smoker")
plt.show()
```

```
sns.relplot(x="total_bill",y="tip",data=df,kind="scatter",row="smoker")
plt.show()
```

```
sns.relplot(x="total_bill", y="tip", data=df, kind="scatter", col="smoker", row="time")
plt.show()
```

```
sns.relplot(x="total_bill",y="tip",data=df,kind="scatter",col="day",col_wrap=2)
plt.show()
```

```
sns.relplot(x="total_bill",y="tip",data=df,kind="scatter",col="day",col_wrap=2,col_order=["Thur","Fri","Sat","Sun"])
plt.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt
sns.relplot(x="total_bill",y="tip",data=df,kind="scatter",col="day",col_wrap=2, col_order=["Thur","Fri","Sat","Sun"])
plt.show()
```

### Scatter plot overview
Show relationship between two quantitative variables
We've seen:
* Subplots (col and row)
*Subgroups with color (hue)
New Customizations:
Subgroups with point size and style
Changing point transparency
Use with both scatterplot() and relplot()

```
sns.relplot(x="total_bill",y="tip",data=df,kind="scatter",size="size")
plt.show()
```

```
sns.relplot(x="total_bill", y="tip",data=df,kind="scatter",size="size",hue="size")
plt.show()
```

```
sns.relplot(x="total_bill",y="tip",data=df,kind="scatter",hue="smoker",style="smoker")
plt.show()
```

```
# Set alpha to be between 0 and 1
sns.relplot(x="total_bill",y="tip",data=df,kind="scatter",alpha=0.4)
plt.show()
```

### What are line plots?
Two types of relational plots: scatter plots and lineplots
####Scatter plots
* Each plot point is an independentobservation

#### Line plots
* Each plot point represents the same "thing",typically tracked over time


```
sns.relplot(x="tip", y="total_bill",data=df,kind="line")
plt.show()
```

```
sns.relplot(x="size", y="tip",data=df,kind="line",style="day",hue="day")
plt.show()
```

```
sns.relplot(x="size", y="tip",data=df,kind="line",style="day",hue="day",markers=True)
plt.show()
```

```
sns.relplot(x="size", y="tip",data=df,kind="line",style="day",hue="day",markers=True,dashes=False)
plt.show()
```
