```
from numpy import arange, sin, pi
```

```
x = arange(0.0, 4*pi, 0.01)
y = sin(x)
```

Grafica sencilla

```
plt.plot(x, y)
```

Configuracion de la grafica y su plano

```
plt.grid(axis='both') #plt.grid -> configuracion de cuadricula
plt.ylabel('Y')
plt.xlabel('X')
plt.legend #a√±adir legenda
```

Resultado

```
plt.grid(axis='both') #plt.grid -> configuracion de cuadricula
plt.ylabel('Y') #etiquetas en Y
plt.xlabel('X') #etiquetas en X
plt.plot(x, y, label="Function f(x)") #ploteamos y a√±adimos la legenda
plt.legend() 
```

```
plt.grid(axis='both') #plt.grid -> configuracion de cuadricula
plt.ylabel('Y') #etiquetas en Y
plt.xlabel('X') #etiquetas en X
plt.plot(x, y, 'red', label="Function f(x)") #ploteamos y a√±adimos la legenda
plt.legend() 
```

```
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots()  # Create a figure and an axes.

ax.plot(x, x, label='lineal')  # Plot some data on the axes.
ax.plot(x, x**2, label='cuadratica')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubica')  # ... and some more.

ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
  # Add a title to the axes.
plt.grid(axis='both') # a√±adimos las cuadr√≠culas.
plt.rcParams["figure.figsize"] = (10, 5) #ajustamos el tama√±o de la figura.
ax.legend()  # Add a legend.
```

```
with plt.style.context('dark_background'):
  x = np.linspace(0, 2, 100)

  fig, ax = plt.subplots()  # Create a figure and an axes.

  ax.plot(x, x, 'r', label='lineal')  # Plot some data on the axes.
  ax.plot(x, x**2, 'g', label='cuadratica')  # Plot more data on the axes...
  ax.plot(x, x**3, 'b', label='cubica')  # ... and some more.

  ax.set_xlabel('variable cost')  # Add an x-label to the axes.
  ax.set_ylabel('medical suplies')  # Add a y-label to the axes.
  ax.set_title("Medical Cost")  # Add a title to the axes.
  plt.grid(axis='both') # a√±adimos las cuadr√≠culas.
  plt.rcParams["figure.figsize"] = (14, 7) #ajustamos el tama√±o de la figura.
plt.show()
```

```
plt.annotate('punto minimo', xy=(33, 0.50), xytext=(14, 0.50), arrowprops=dict(facecolor='blue', shrink=2.05)) 
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'g')
    plt.plot(np.sin(np.linspace(0, 3 * np.pi)), 'r')
plt.grid(axis='both')
plt.show()
```

```
plt.annotate('(23,-1)', xy=(33, 0.50), xytext=(23, -0.90)) 
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'g')
    plt.plot(np.sin(np.linspace(0, 3 * np.pi)), 'r')
plt.grid(axis='both')
plt.show()
```
