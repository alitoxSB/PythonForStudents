About trading:

Financial trading is the buying and selling of financial asset.

* Equities
* Bonds
* Forex
* Commodities
* Cryptocurrencies

To make a prot by taking calculated risks.

Long positions: prot from upward price movement

Short positions: prot from downward price movement

Market participants:

* Institutional traders
* Retail traders

#### Trading vs. investing
Trading:

* Shorter holding period

* Focus on short-term trends or price
fluctuations

* Take both long and short positions

#### Investing:
* Longer holding period
* Focus on market fundamentals
* Take mostly long positions

### Different types of traders
Day Trader: holds positions throughout the day but usually not overnight.

Swing Trader: holds positions from a few days to several weeks.

Position Trader: holds positions from a few months to several years.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
bitcoin_data = pd.read_csv("Bitcoin.csv", index_col='Date', parse_dates=True)
print(bitcoin_data.head())
```

### Plot line chart of time series data

```python
with plt.style.context('dark_background'):
  plt.plot(bitcoin_data['Close'], color='y')
  plt.title("Daily close price")
  plt.grid(axis='both')
  plt.rcParams["figure.figsize"] = (14, 12)
plt.show()
```

```python
import plotly.graph_objects as go
# Define the candlestick
candlestick = go.Candlestick(x = bitcoin_data.index, open = bitcoin_data['Open'],high = bitcoin_data['High'],low = bitcoin_data['Low'],close = bitcoin_data['Close'])

# Create a plot
fig = go.Figure(data=[candlestick])
# Show the plot
fig.show()
```

### Google Stocks Data: Calculate for plot the daily returns.

```python
goole_stock_data = pd.read_csv("GOOG.csv", index_col='Date', parse_dates=True)
goole_stock_data
```

```python
# Calculate daily returns
goole_stock_data['daily_return'] = goole_stock_data['Close'].pct_change() * 100
# Plot the data
plt.plot(goole_stock_data['daily_return'])
plt.rcParams["figure.figsize"] = (10, 5)
plt.show()
```

```python
goole_stock_data['daily_return'].hist(bins=100)
plt.show()
```

### Tesla Stocks Data: Calculate for plot the daily returns.

```python
Tesla_stock_data = pd.read_csv("TSLA.csv", index_col='Date', parse_dates=True)
Tesla_stock_data
```

```python
# Calculate daily returns
Tesla_stock_data['daily_return'] = Tesla_stock_data['Close'].pct_change() * 100
with plt.style.context('dark_background'):
  # Plot the data
  plt.title("Daily returns for Tesla stock")
  plt.grid(axis='both')
  plt.plot(Tesla_stock_data['daily_return'],color='r', label="TSLA Stock Daily Returns")
  plt.rcParams["figure.figsize"] = (14, 7)
  plt.legend() 
plt.show()
```

```python
with plt.style.context('dark_background'):
  Tesla_stock_data['daily_return'].hist(bins=100,color='r')
plt.show()
```

### The bt package
A exible framework for dening and backtesting trading strategies.

* Strategy: a method of buying and selling nancial assets based on predened rules.

* Strategy backtesting: a way to assess the eectiveness of a strategy by testing it on historical data

### The bt process
* Step 1: Get the historical price data
* Step 2: Dene the strategy
* Step 3: Backtest the strategy with the data
* Step 4: Evaluate the result
