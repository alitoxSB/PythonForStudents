### Extraccion de listas de articulos


```python
def obtener_notas(soup):
    lista_notas = []

    ##obtencion de articulo promocionado
    featured_article = soup.find('div', attrs={'class':'featured-article_container'}) 
    if featured_article:
      lista_notas.append(featured_article.a.get('href'))

    ##obtencion del listado de articulo
    article_list = soup.find('ul', attrs={'class':'article-list'})
    for article in article_list.find_all('li'):
      if article.a:
        lista_notas.append(article.a.get('href'))
    
    return lista_notas
```

```python
lista_notas = obtener_notas(s_seccion)
lista_notas
```

### Descargar contenido

```python
url_nota = lista_notas[0]
url_nota
```

```python
try:
  nota = requests.get(url_nota)
  if nota.status_code == 200:
    s_nota = BeautifulSoup(nota.text, 'lxml')
  ## extraemos el titulo
  title = s_nota.find('h1', attrs={'class': 'article-title'})
  print(title.text)

  ##Extraer la fecha
  fecha = s_nota.find('span', attrs={'pubdate':'pubdate'}).get('datetime')
  print(fecha)

  ##Extraer el autor
  autor = s_nota.find('div', attrs={'class':"article-author"}).a
  print(autor.text)

  ##Extraer volantado
  volantado = s_nota.find('h2', attrs={'class':"article-prefix"})
  print(volantado.get_text())

except Exception as e:
    print('error: ')
    print(e) 
    print('\n')
```

### Extraccion de contenido multimedia

```python
media = s_nota.find('div', attrs={"class": "article-main-media-image"})
media.find_all('img') #el find all es para quedarnos con todas las imagenes
imagenes = media.find_all('img')
```

```python
if len(imagenes)==0:
  print('no se encontraron imagenes')
else:
  imagen = imagenes[-1]
  img_src = imagen.get('data-src')
  print(img_src)
```

```python
img_req = requests.get(img_src)
img_req.status_code
```
