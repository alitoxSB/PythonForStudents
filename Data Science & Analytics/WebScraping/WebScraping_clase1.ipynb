```python
import requests
```

```python
url = 'https://www.pagina12.com.ar/'
```

```python
p12 = requests.get(url)  ##De esta forma descargamos la pagina web
```

```python
p12.status_code ##Verificamos el statuscode.
```

```python
print(p12.text) ##imprmimos el texto
```

```python
p12.content ##Entender el contenido
```

```python
p12.headers
```

```python
p12.request.headers ##encabezados con la solicitud que hacemos
```

```python
p12.request.method ##saber que metodo estamos utilizando
```

```python
p12.request.url
```

```python
from bs4 import BeautifulSoup
```

```python
s = BeautifulSoup(p12.text, 'lxml') #separa el texto largo en partes pequenas para su manejo
```

```python
print(s.prettify()) ##Estrutura de la pagina
```

```python
secciones = s.find('ul', attrs={'class':'hot-sections'}).find_all('li') ##imprimimos el menu
secciones
```

```python
seccion = secciones[1] ## obtenemos la primera seccion
```

```python
seccion.a ## nos quedamos con el atributo a
```

```python
seccion.a.get("href") ##obtenemos el link
```

```python
seccion.a.get_text() ## obtenemos el texto
```

```python
links_secciones = [seccion.a.get('href') for seccion in secciones]
```

```python
links_secciones
```

```python
sec = requests.get(links_secciones[0])
sec.status_code
```

```python
s_section = BeautifulSoup(sec.text, 'lxml')
s_section
```

```python
print(s_section.prettify())
```
